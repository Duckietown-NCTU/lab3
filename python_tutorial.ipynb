## example for function evaluation 

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
```

```python
xx, yy = np.meshgrid(x, y, sparse = True)
#print xx
#print yy
z = np.sin(xx*2+yy*2)/(xx*2+yy*2)
h = plt.contourf(x, y, z)
plt.show()
```
